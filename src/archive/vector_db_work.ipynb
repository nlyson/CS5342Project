```python
import chromadb
import PyPDF2
import nltk

```

```python
test_file = '../docs/ComputerNetworkingTopDownApproach.pdf'

sentences = []

with open(test_file, 'rb') as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    print(f'There are [{len(pdf_reader.pages)}] pages')
    for page in pdf_reader.pages:
        sentences += nltk.sent_tokenize(page.extract_text())

print(f'There are [{len(sentences)}] sentences.')
sentences

        
```
